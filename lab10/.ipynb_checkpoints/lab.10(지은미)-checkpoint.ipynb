```python
# 1번 (Forward propagation)
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

i=np.array([0.9,0.1,0.8])
w_input = np.array([[0.9,0.3,0.4],[0.2,0.8,0.2],[0.1,0.5,0.6]])
w_out=np.array([[0.3,0.7,0.5],[0.6,0.5,0.2],[0.8,0.1,0.9]])
x=np.dot(w_input,i)
print(x)

o_hidden = sigmoid(x)
print(o_hidden)

x_out = np.dot(w_out,o_hidden)
print(x_out)

o_out = sigmoid(x_out)
print(o_out)
```

```python
# 2번 (Backward propagation)
w1 = np.array([[2.0,1.0],[3.0,4.0]])
w2 = np.array([[3.0,1.0],[2.0,7.0]])               
e1 = 0.8 * (w1[0][0] / (w1[0][0] + w1[1][0])) + 0.5 * (w1[0][1] / (w1[0][1] + w1[1][1]))
e2 = 0.8 * (w1[1][0]/(w1[1][0]+w1[0][0]))+0.5*(w1[1][1]/(w1[1][1]+w1[0][1]))
print(e1)
print(e2)


```

```python
# 3번 (Weight Update)
current_w11 = 2.0
w1 = np.array([[2.0,1.0],[3.0,4.0]])
w2 = np.array([[3.0,1.0],[2.0,7.0]])
o=np.array([0.4,0.5])
e1=0.8
e2=0.5
w11=-e1*sigmoid((w1[0][0]*o[0])+(w1[1][0]*o[1]))*(1-sigmoid((w1[0][0]*o[0])+(w1[1][0]*o[1])))*o[0]
print(w11)
study = 0.1
w11=current_w11-(w11*study)
print(w11)
```
