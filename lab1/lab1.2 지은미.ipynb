```python
list1=['a','b','c']
list2=[1,[2,'x',3,'y'],'z']

dictionary=dict()

for i in range(3):
    key=list1[i]
    value=list2[i]
    if key not in dictionary:
        dictionary.update({key:value})
print(dictionary)

for key in dictionary:
    print(key,":",dictionary[key])

```
