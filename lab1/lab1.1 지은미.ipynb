```python
fruit1=["apple"]
fruit2=["banana"]
fruit3=["orange"]

list=fruit1+fruit2+fruit3


for i in range(len(list)):
    print(i,":",list[i])
    
file=open('lab1.1.csv','w')
out=",".join(list)
file.write(out)
file.close()

file=open('lab1.1.csv','r')
out=file.read()
print(out)
file.close()
```
