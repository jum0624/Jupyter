```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

def forward_propagation(i,w):
    x=np.dot(w,i)
    o=sigmoid(x)
    return o

input_value = np.array([1.0,0.5], ndmin=2).T
w=np.array([[0.9,0.3],[0.2,0.8]], ndmin=2)

output = forward_propagation(input_value,w)
print(output)
```
