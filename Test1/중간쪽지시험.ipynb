```python
#3
class Father:
    def handsome(self):
        print("잘생기셨습니다!")

class Son(Father):
    def tall(self):
        print("키가 큽니다")

class Daughter(Father):
       
    def handsome(self):
        print("예쁩니다")
    

son=Son()
son.tall()
son.handsome()


girl=Daughter()
girl.handsome()
```

```python
#5
import numpy as np
onehots = dict()
onehots['cat'] = [1,0,0,0]
onehots['the'] = [0,1,0,0]
onehots['dog'] = [0,0,1,0]
onehots['sat'] = [0,0,0,1]
onehots
sentence = ['the','cat','sat']
x = np.array([0, 0, 0, 0])
for name in sentence:
    for result, i in enumerate(onehots[name]):
        if i == 1:
            x[result] = 1
print("Sentence Encoding: " + str(x))
```
